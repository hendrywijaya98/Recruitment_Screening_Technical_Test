Data Cleaning Shipping City

```python
# remove Kab. and Kota
raw_clothes['shipping_city'] = raw_clothes['shipping_city'].str.replace('Kota ','')
raw_clothes['shipping_city'] = raw_clothes['shipping_city'].str.replace('Kabupaten','')
raw_clothes['shipping_city'] = raw_clothes['shipping_city'].str.replace('Kab','')
# remove administrasi then adm
raw_clothes['shipping_city'] = raw_clothes['shipping_city'].str.replace('Administrasi','')
raw_clothes['shipping_city'] = raw_clothes['shipping_city'].str.replace('Adm','')
# remove dot symbol with regex function and white space with strip
raw_clothes['shipping_city'] = raw_clothes['shipping_city'].str.replace('\.','', regex=True)
raw_clothes['shipping_city'] = raw_clothes['shipping_city'].str.strip()
raw_clothes['shipping_city'] = raw_clothes['shipping_city'].str.replace('Ketapant','Ketapang')

# raw_clothes[raw_clothes['shipping_city'].str.contains('Jakarta')]['shipping_city'].unique()
# print(len(raw_clothes['shipping_city'].unique()))
# raw_clothes['shipping_city'].unique()
```

Data Cleaning numeric object to truly numeric column

then change all minus contained numericv value into absolute

```python
# gross sales columns
raw_clothes['gross_sales'] = raw_clothes['gross_sales'].str.replace('\.','', regex=True)
raw_clothes['gross_sales'] = raw_clothes['gross_sales'].astype(float).fillna(0)
# print(raw_clothes['gross_sales'].unique())

# discoutns columns
raw_clothes['discounts'] = raw_clothes['discounts'].str.replace('-','', regex=True)
raw_clothes['discounts'] = raw_clothes['discounts'].str.replace('\.','', regex=True)
raw_clothes['discounts'] = raw_clothes['discounts'].astype(float).fillna(0)
# print(raw_clothes['discounts'].unique())

# returns columns
raw_clothes['returns'] = raw_clothes['returns'].str.replace('-','', regex=True)
raw_clothes['returns'] = raw_clothes['returns'].str.replace('\.','', regex=True)
raw_clothes['returns'] = raw_clothes['returns'].astype(float).fillna(0)
# print(raw_clothes['returns'].unique())

# net_sales columns
raw_clothes['net_sales'] = raw_clothes['net_sales'].str.replace('-','', regex=True)
raw_clothes['net_sales'] = raw_clothes['net_sales'].str.replace('\.','', regex=True)
raw_clothes['net_sales'] = raw_clothes['net_sales'].astype(float).fillna(0)
# print(raw_clothes['net_sales'].unique())

# total_sales columns
raw_clothes['total_sales'] = raw_clothes['total_sales'].str.replace('-','', regex=True)
raw_clothes['total_sales'] = raw_clothes['total_sales'].str.replace('\.','', regex=True)
raw_clothes['total_sales'] = raw_clothes['total_sales'].astype(float).fillna(0)
# print(raw_clothes['total_sales'].unique())
```

```python
def product_category(product_title):
    category = ''
    if 'Blouse' in product_title:
        category = 'Blouse'
    elif 'Dress' in product_title:
        category = 'Dress'
    elif 'Tunic' in product_title:
        category = 'Tunic'
    elif 'Outer' in product_title:
        category = 'Outer'
    elif 'Pants' in product_title:
        category = 'Pants'
    elif 'Scarf' in product_title:
        category = 'Scarf'
    elif 'Skirt' in product_title:
        category = 'Skirt'
    elif 'Pashmina' in product_title:
        category = 'Pashmina'
    elif 'Shirt' in product_title:
        category = 'Shirt'
    elif 'Brooch' in product_title:
        category = 'Brooch'
    elif 'Prayer' in product_title:
        category = 'Prayer'
    elif 'Skirt' in product_title:
        category = 'Skirt'
    return category

# raw_clothes['product_category'] = raw_clothes['product_title'].apply(product_category)
raw_clothes.insert(1,'product_category', raw_clothes['product_title'].apply(product_category))
```
